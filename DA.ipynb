### Data Cleaning

```python
#create a dataset with all the types of errors
student_data = {"name": ["Velan", "Rajesh", "Divya", "sakthi", "Divya"],
                "Roll_no": [12,       23,        17,        50,      17],
                "age": [ "nineteen", np.nan,    21,         190,      21],
                "marks": [  96,        89,     np.nan,      97,      np.nan]}
```

```python
student_df = pd.DataFrame(student_data)
student_df
```

```python
student_df.info()
```

Issues in the student dataframe
- Duplicate values
- error values
- Outlier
- Missing values

```python
#Remove duplicate values
student_df.drop_duplicates(inplace = True)
student_df
```

```python
#fix the error value in the age column and at the index 0
student_df["age"][0] = 19
```

```python
student_df
```

```python
student_df["age"][3] = 19
```

```python
student_df
```

```python
student_df["marks"][2] = 0
student_df
```

```python
student_df["age"] = student_df["age"].convert_dtypes(convert_integer=True)
student_df.info()
```

```python
#Fill the missing value with the mean of the column
student_df.fillna(int(student_df["age"].mean()),inplace=True)
student_df
```
